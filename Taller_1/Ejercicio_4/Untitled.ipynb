```python
import pandas_datareader.data as web
import datetime

f = web.DataReader('USD000UTSTOM', 'moex', start='2017-01-01', end='2019-12-01')
f.tail()
```

```python
data_OC = f.filter(regex = '(OPEN|CLOSE)')
data_OC['return'] = data_OC['OPEN']-data_OC['CLOSE']
data_OC.head()
```

```python
import pandas_montecarlo
lower_bound = -0.2
upper_bound = 0.05
bust_values = [-0.1, 0.04, 0.0, -0.132, 0.023]
simulations = []
n_simulations = 10
sims_v = 11
index = 0
mc = data_OC['return'].montecarlo(sims=10, bust=-0.2, goal=0.1)
mc2 = data_OC['return'].montecarlo(sims=10, bust=0.05, goal=0.1)
```

```python
import plotly.graph_objects as go
mc.plot(title="Simulaciones de variacin en el incremento con un bust de: " + str(-0.2))
mc2.plot(title="Simulaciones de variacin en el incremento con un bust de: " + str(0.05))

```
